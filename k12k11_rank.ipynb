```python
def nyse_day_delta(day_delay, date_start, day_delta=30):
    """The function takes a dataframe with a single column. Each row is a
        list-of-lists of symbols. The function flattens the list-of-list in each
        row and return it as list of symbols.

        Args:
            date_start(str): date in 'yyyy-mm-dd' format
            day_delay(int): delay in trading days
            day_delta(int): calendar day increase from date_start

        Return:
            list: flattened list of symbols,
            e.g. ['NIO', 'TSLA', 'SE', 'IRTC', 'W']    
        import datetime
        import pandas_market_calendars as mcal
        from datetime import datetime as dt
    """
    import datetime
    import pandas_market_calendars as mcal
    from datetime import datetime as dt

    # convert to datetime object
    dt_start = dt.strptime(date_start, '%Y-%m-%d')
    dt_end = dt_start + datetime.timedelta(days=day_delta)
    start_date = dt_start.strftime('%Y-%m-%d')  # str
    end_date = dt_end.strftime('%Y-%m-%d')  # str
    nyse = mcal.get_calendar('NYSE')
    date_nyse = nyse.schedule(start_date=start_date, end_date=end_date)
    # convert index from datetime to string
    date_index = mcal.date_range(date_nyse, frequency='1D').strftime('%Y-%m-%d')
    date_start_loc = date_index.get_loc(start_date)
    date_end_loc = date_start_loc + day_delay
    date_end = date_index[date_end_loc]

    return date_end


# date_start = df_sym_best_pct_rank.iloc[0]['date_end']
# # date_start = '2020-12-24'
# print(f'date_start: {date_start}, type(date_start): {type(date_start)}')
# day_delay = 1
# date_end = nyse_day_delta(date_start=date_start, day_delay=day_delay)
# print(date_start, day_delay, date_end, type(date_end))
```
